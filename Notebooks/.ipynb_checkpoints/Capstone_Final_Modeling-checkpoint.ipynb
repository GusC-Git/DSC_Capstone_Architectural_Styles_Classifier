## Organizing Data
Because I am not on kaggle for this part, I need to redo some small processing I performed for the data. I have added if statements to prevent these changes from every running a second time.

```python
#Creates a file that will contain all the images in the dataset if not already present
if 'allimages' not in os.listdir(file_root_dir):
  os.makedirs(file_root_dir+'allimages')

  #Creates a copy of every image in the allimages file
  for period in periods:
      styles_in_period = os.listdir(data_root_dir+period)
      for style in styles_in_period:
          src = data_root_dir + period + '/' + style
          allFileNames = os.listdir(src)
          allImgNames = [src+'/'+ name for name in allFileNames]
          for name in allImgNames:
              shutil.copy(name, file_root_dir +'allimages')

else:
  print('folder already exists and contains said images')
```

```python
imageFolder = file_root_dir+'allimages/'
listImages = os.listdir(file_root_dir+'allimages/')

#Checks if the image has exif data. If it does not, image is added into an errored images list
errored_images = []
for img in listImages:
    imgPath = os.path.join(imageFolder,img)
            
    try:
      imag = Image.open(imgPath)
      exif_data = imag._getexif()
        
    except:
      print("Error on image: ", img)
      errored_images.append(img)
```

```python

val_ratio =0.1
test_ratio = 0.1
```

```python
if not len(os.listdir(file_root_dir+'train/21st_Century')):
  for period in periods:
    styles_in_period = os.listdir(data_root_dir+period)
    total_count = 0
    train_count = 0
    val_count = 0
    test_count = 0
    for style in styles_in_period:
      src = data_root_dir + period + '/' + style
      allFileNames = os.listdir(src)
      allFileNames = [FileName for FileName in allFileNames if FileName not in errored_images]
      np.random.shuffle(allFileNames)
      train_FileNames, val_FileNames, test_FileNames = np.split(np.array(allFileNames),
                                                            [int(len(allFileNames)* (1 - (val_ratio + test_ratio))),
                                                             int(len(allFileNames)* (1 - test_ratio))])
      train_FileNames = [src+'/'+ name for name in train_FileNames.tolist()]
      val_FileNames = [src+'/' + name for name in val_FileNames.tolist()]
      test_FileNames = [src+'/' + name for name in test_FileNames.tolist()]
        
      total_count += len(allFileNames)
      train_count += len(train_FileNames)
      val_count += len(val_FileNames)
      test_count += len(test_FileNames)
    
      for name in train_FileNames:
        shutil.copy(name, file_root_dir +'train/' + period)

      for name in val_FileNames:
        shutil.copy(name, file_root_dir +'val/' + period)

      for name in test_FileNames:
        shutil.copy(name, file_root_dir +'test/' + period)
            
    print('Total images of {}: '.format(period), total_count, )
    print('Training: ', train_count)
    print('Validation: ', val_count)
    print('Testing: ', test_count)
else:
  print('training data already present. Check google drive')
```

## Final Model
I did experiementing with models within kaggle and came up with a model structure that performed the best for me after a few iterations. This is the structure and these are the results I will record as my final model.

```python
%matplotlib inline
def show_acc_loss_graphs(modelhistory):
    """
    Takes in the history of a model and produces graphs that show the accuracy change and the loss on both train and validation sets over epochs
    
    Inputs
    ------
    modelhistory: return output of keras model.fit.
    
    From this the accuracy and loss is gathered from for plotting.
    """
    acc = modelhistory.history['acc']
    val_acc = modelhistory.history['val_acc']
    loss = modelhistory.history['loss']
    val_loss = modelhistory.history['val_loss']
    epochs = range(len(acc))

    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.plot(epochs, acc, 'go', label= 'Training Acc')
    plt.plot(epochs, val_acc, 'g', label= 'Validation Acc')
    plt.title('Training and Validation Accuracy')
    plt.legend()
    plt.figure()

    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.plot(epochs, loss, 'go', label='Training loss')
    plt.plot(epochs, val_loss, 'g', label='Validation loss')
    plt.title('Training and validation loss')
    plt.legend()
    plt.show()
    return
```

```python
train_dir = file_root_dir + 'train/'
val_dir = file_root_dir + 'val/'
test_dir = file_root_dir + 'test/'

train_datagen = ImageDataGenerator(rescale=1./255, shear_range=0.2, zoom_range=[0.8,1.2], brightness_range= (0.6, 0.9), fill_mode='nearest')
val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(train_dir,\
                                                    target_size=(260, 200),\
                                                    batch_size= 40,\
                                                    class_mode='categorical')

validation_generator = val_datagen.flow_from_directory(val_dir,\
                                                      target_size=(260, 200),\
                                                      batch_size=40,\
                                                      class_mode='categorical')

test_datagen = ImageDataGenerator(rescale=1./255)

test_generator = test_datagen.flow_from_directory(test_dir, 
                                                  target_size=(260, 200), 
                                                  batch_size=40, 
                                                  class_mode='categorical')
```

```python
model = models.Sequential()

model.add(layers.Conv2D(64, (5,5), activation= 'relu',padding= 'same', input_shape= (260,200,3)))
model.add(layers.MaxPooling2D((2,2)))


model.add(layers.Conv2D(128, (3,3), padding= 'same', activation= 'relu'))
model.add(layers.Conv2D(128, (3,3), padding= 'same', activation= 'relu'))
model.add(layers.MaxPooling2D((2,2)))


model.add(layers.Conv2D(256, (3,3), padding= 'same', activation ='relu'))
model.add(layers.Conv2D(256, (3,3), padding= 'same', activation ='relu'))
model.add(layers.MaxPooling2D((2,2)))

model.add(layers.Conv2D(512, (3,3), padding= 'same', activation ='relu'))
model.add(layers.Conv2D(512, (3,3), padding= 'same', activation ='relu'))
model.add(layers.MaxPooling2D((2,2)))

model.add(layers.Conv2D(512, (3,3), padding= 'same', activation ='relu'))
model.add(layers.Conv2D(512, (3,3), padding= 'same', activation ='relu'))
model.add(layers.MaxPooling2D((2,2)))

model.add(layers.Flatten())
model.add(layers.Dense(100, activation='relu'))
model.add(layers.Dense(9, activation='softmax'))
```

```python
myoptimizer = keras.optimizers.Adam(learning_rate= 3e-4)
model.compile(optimizer= myoptimizer, loss='categorical_crossentropy', metrics= ['acc'])
model.summary()
```

```python
history = model.fit(train_generator,\
                   epochs=10,\
                   validation_data=validation_generator)
```

```python
show_acc_loss_graphs(modelhistory=history)
```

In previous runs past this number of epochs the model began overfitting onto the training data and validation accuracy dropped. Within 10 epochs I was able to achieve 50% accuracy on my data.

```python
test_gen_for_preds = test_datagen.flow_from_directory(test_dir, 
                                                  target_size=(260,200), 
                                                  batch_size=40, 
                                                  class_mode='categorical',
                                                     shuffle=False)
y_preds = model.predict(test_gen_for_preds)

k_range = np.linspace(1,9,9)
acc_list = []
for k in k_range:
    
    acc = np.sum(tf.math.in_top_k(predictions=y_preds, targets=test_gen_for_preds.classes, k=k))/1248
    acc_list.append(acc)
    
```

```python
plt.figure(figsize=(7,7))
plt.xlabel('Number of top features',{'fontsize':12})
plt.ylabel('Accuracy Score',{'fontsize':12})
plt.title("Accuracy within top K features", {'fontsize':15})
plt.axvline(x=3, ymin=0, ymax=1, c='red',label="Top 3 feature cutoff")
plt.plot(k_range, acc_list,linewidth=3, label ="Accuracy per within top k features")
plt.legend()
plt.show()
```

When looking at the model, we achieve 80% accuracy within the top 3 features This is pretty good considering the limitations experienced with data.

```python
cm =confusion_matrix(test_generator.classes, np.argmax(y_preds, axis=1))
plt.figure(figsize=(9,9))
sns.heatmap(cm, xticklabels= list(test_generator.class_indices.keys()), yticklabels= list(test_generator.class_indices.keys()), annot= True)
plt.show()
```

From the heatmap it can be deduced that the model was best able to classify architecture belonging to Modernism and Revivalism. This heatmap also shows that our model misclassified a lot of 21st Century architecture as modernism. This makes some sense, as 21st Century incorporated post-modernist designs and by nature, is the style that follows this period. This trend can be spotted in many different parts throughout the confusion matrix: styles that come before or supercede one another are more likely to misclassified for one another. For every style period except 21st Century, the model was able to predict the correct class more than it would predict a different, incorrect one.

# View convolutional layers

Now that I have my final model, I will visualize some of the activation layers so that we see what is happening within the model. I will visualize the first layer in its entirety and then just one slice from the following layers

```python
model = keras.models.load_model(file_root_dir+'finalcnnmodel.h5')
```

```python
img_path = file_root_dir + 'train/Modernism/50417.jpg'

#Grab image and do original model preprocessing
img = keras.preprocessing.image.load_img(img_path, target_size= (260, 200))
img_tensor = keras.preprocessing.image.img_to_array(img)
img_tensor = np.expand_dims(img_tensor, axis=0)
img_tensor /= 255.

print(img_tensor.shape)
plt.figure(figsize=(6.5, 5))
plt.imshow(img_tensor[0])
plt.show()
```

```python
#Extract layer outputs
layer_outputs = [layer.output for layer in model.layers[:14]]

#Making a model to display the feature maps
activation_model = models.Model(inputs=model.input, outputs= layer_outputs)
```

```python
activations = activation_model.predict(img_tensor)

first_layer_activation = activations[0]

plt.matshow(first_layer_activation[0,:,:, 42], cmap='rainbow')
plt.show()
```

```python
#Plots all of the different activations within the first layer of the network.
fig, axes = plt.subplots(8, 8, figsize=(24, 24))
for i in range(64):
  row = i//8
  column = i%8
  ax = axes[row, column]
  first_layer_activation = activations[0]
  ax.matshow(first_layer_activation[0,:,:, i], cmap='rainbow')
```

This is all 64 images produced in the first convolutional layer. As we can see, The model paid a lot of attendtion where edges are located, and in some images, separated foreground from background. While a Convolutional Neural Network is kind of a black box in the sense that we don't know exactly what it does with these images, we are able to see all of the images that it is able to produce.

```python
# Visualizing a singualr layer across the entire model 
# The last two positions will be missing because we only have 14 graphs
fig, axes = plt.subplots(4, 4, figsize=(16,18))

layer_names = []
for layer in model.layers[:14]:
  layer_names.append(layer.name)

for i in range(14):
  row = i//4
  column = i%4
  ax = axes[row, column]
  current_layer = activations[i]
  ax.matshow(current_layer[0, :,:, 11], cmap='rainbow')
  ax.xaxis.set_ticks_position('bottom')
  ax.set_title(layer_names[i])
```

Last two are purposefully left blank so that I may use a 4x4 square instead of some 2x7 shape. This visualizes all images within channel 11 of our model at different activation periods. Latter convolutions begin to give human viewers a less intuitive insight of what's happening within the model.
